```python
from datasets import load_dataset
from itertools import product
```

```python
mapping = {''.join(p): i for i, p in enumerate(product('ACGT', repeat=3))}

def seq_to_trinucleotides(row):
    seq = row['sequence']
    if len(seq) == 606:
        motif_len = 6
    elif len(seq) == 603:
        motif_len = 3
    upstream = seq[:300]
    downstream = seq[300+motif_len:]
    upstream_nuc = list(map(mapping.get, [upstream[i:i+3] for i in range(len(upstream)-2)]))
    downstream_nuc = list(map(mapping.get, [downstream[i:i+3] for i in range(len(downstream)-2)]))
    return {'upstream': upstream_nuc, 'downstream': downstream_nuc}
```

```python
dataset = load_dataset('dvgodoy/DeepGSR_sequences')
dataset = dataset.filter(lambda row: 'N' not in row['sequence'])
dataset = dataset.map(seq_to_trinucleotides)
dataset = dataset.select_columns(['id', 'organism', 'motif', 'label', 'signal', 'upstream', 'downstream'])
```

```python
dataset.push_to_hub('dvgodoy/DeepGSR_trinucleotides')
```
